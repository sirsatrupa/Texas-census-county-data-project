```python
import pandas as pd
```

```python
df = pd.read_csv("5. County Stats.csv")
```

```python
df.head()
```

```python
def cut_county_name(s):
    return s[:-7]

df["County"] = df["County"].map(cut_county_name)
```

```python
pop_df = df[["County", "Total Population"]].copy()
```

```python
pop_df.head()
```

```python
pop_df.to_csv("County Pop Data.csv")
```
